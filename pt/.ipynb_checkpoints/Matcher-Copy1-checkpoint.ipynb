```python
import pandas as pd
from shapely import Polygon, Point, buffer, LineString, union_all, normalize
```

```python
df_routes
```

```python
def parse_linestr(linestring):
    coord_list = linestring[11:][:-1].split(",")
    coord_list = [x.strip() for x in coord_list]
    point_list_out = []
    longitudes = []
    latitudes = []
    for coord in coord_list:
        lon, lat = coord.split(" ")
        longitudes.append(float(lon))
        latitudes.append(float(lat))
        point_list_out.append((lon, lat))

    return point_list_out

def create_buffer(polyline, band_width=0.0001):
    ls = LineString(polyline)
    buffer_polygon = buffer(ls, distance=band_width, cap_style="flat")
    return buffer_polygon

df_links = pd.read_json("links.json")
df_links['buffer_polygon'] = df_links['polyline'].apply(create_buffer)
df_links
```

```python
df_routes = pd.read_csv("final_result1.csv", index_col=0)
```

```python
def get_routes_within_polygon(polygon: Polygon):
    filters = df_routes['polyline'].apply(parse_linestr).apply(LineString).apply(polygon.intersects)
    return df_routes[filters][['route_number', 'from_stop', 'to_stop']].values
```

```python
from tqdm import tqdm
tqdm.pandas()

df_links['routes_within_buffer'] = df_links['buffer_polygon'].progress_apply(get_routes_within_polygon)
```

```python
df_links.to_csv("links_with_adjacent_pt_routes.csv")
```

```python
def pa(l):
    return ['-'.join([str(n) for n in x.tolist()]) for x in l]
df_links['routes_within_buffer_2'] = df_links['routes_within_buffer'].apply(pa)
def revmap(row):
    values = row[['route_number', 'from_stop', 'to_stop']].values
    tup = '-'.join([str(n) for n in values.tolist()])

    bools = []
    for i, row in df_links.iterrows():
        if tup in row['routes_within_buffer_2']:
            bools.append(True)
        else:
            bools.append(False)
    
    r = df_links[bools]['link_id'].values
    return r
    
df_routes['links_within_buffer'] = df_routes.progress_apply(revmap, axis=1)
```

```python
df_routes['links_within_buffer'] = df_routes['links_within_buffer'].apply(lambda x: ','.join(x))
df_routes.to_csv("pt_routes_with_adjacent_links_2.csv")
```

```python
df_o = df_routes[df_routes.route_number == "3"].copy()
df_o['links_within_buffer'] = df_o['buffer_polygon'].apply(get_links_within_polygon).apply(lambda x: ','.join(x))
df_o
```
