```python
df = pd.read_csv("dp.csv", index_col=0)
```

```python
df
```

```python
melb_map = folium.Map(location=[-37.82003422345045, 145.01656189422235], zoom_start=12, tiles="cartodb positron")
colors = ['red', 'yellow', 'green', 'blue', 'purple']
for i, row in df.iterrows():
    folium.CircleMarker(location=[row.yj, row.xj],
                        radius=2,
                        color=colors[row.preds],
                        weight=5).add_to(melb_map)
melb_map
```
